# 판다스 입문 

## 1. 데이터과학자가 판다스를 배우는 이유

 실제 데이터 분석 업무의 80~90%는 데이터를 수집하고 정리하는 일 차지한다고 한다. 판다스 라이브러리는 데이터를 수집하고 정리하는 데 최적화된 도구이다. 오픈소스라서 무료이고  
가장 배우기 쉬운 프로그래밍 언어 중의 하나인 파이썬을 기반으로 하기 때문에 배우기 쉽다. 판다스를 배우면 데이터과학의 80~90%의 업무를 처리할 수 있고, 데이터과학자에게 필요한  
기본적이면서도 아주 중요한 도구를 갖추게 된다.

## 2. 판다스 자료구조

분석을 위해 다양한 소스로부터 수집하는 데이터는 형태나 속성이 매우 다양하다. 이를 컴퓨터가 이해할 수 있도록 동일한 형식을 갖는 구조로 통합할 필요가 있다.  
이를 위해 판다스는 시리즈와 데이터프레임이라는 구조화된 데이터 형식을 제공한다.  
판다스의 1차적인 목적은 서로 다른 여러 가지 유형의 데이터를 공통의 포멧으로 정리하는 것이다.  
***판다스는 어떤 내용으로 구성되었을까?***  
판다스 라이브러리는 여러 종류의 클래스와 다양한 내장 함수로 구성된다. 시리즈와 데이터프레임 클래스의 속성과 메소드를 잘 이해하면 판다스를 자유자재로 다루는 데 어려움이 없을 것이다.  
내장 함수로는 `Series(), DataFrame(), read_csv(), read_excel()` 등이 있다.

### 2-1 시리즈

시리즈는 데이터가 순차적으로 나열된 1차원 배열의 형태를 갖는다. 인덱스는 데이터 값과 일대일 대응이 된다.  
이처럼 시리즈의 인덱스는 데이터 값의 위치를 나타내는 이름표역할을 한다.

- 시리즈 만들기

````python
# <예제 1-1>
# 딕셔너리 -> 시리즈 변환
import pandas as pd

dict_data = {'a': 1, 'b': 2, 'c': 3}

sr = pd.Series(dict_data)

print(type(sr))

print(sr)
````

- 인덱스 구조

````python
# <예제 1-2>
# 시리즈 인덱스
import pandas as pd

list_data = ['2019-01-02', 3.14, 'ABC', 100, True]
sr = pd.Series(list_data)
print(sr)

# 인덱스 배열은 변수 idx에 저장
idx = sr.index
print(idx)

# 데이터 값 배열은 변수 val에 저장
val = sr.values
print(val)

# 시리즈 배열을 구성하는 원소의 자료형
print(sr.dtype)

# 시리즈 배열의 크기
print(len(sr))

# 시리즈 배열의 형태
print(sr.shape)

# 시리즈 배열의 차원
print(sr.ndim)
````

- 원소 선택

````python
# <예제 1-3>
# 시리즈 원소 선택
import pandas as pd

# 튜플을 시리즈로 변환(인덱스 옵션 지정)
tup_data = ('영인', '2010-05-01', '여', True)
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])
print(sr)
print('\n')

# 원소를 1개 선택
print(sr.iloc[0])    # sr의 1번째 원소를 선택(정수형 위치 인덱스)
print(sr.loc['이름'])    # '이름' 라벨을 가진 원소를 선택(인덱스 이름)
print('\n')

# 여러 개의 원소를 선택(인덱스 리스트 활용)
print(sr.iloc[[1, 2]])
print('\n')
print(sr.loc[['생년월일', '학생여부']])
print('\n')

# 여러 개의 원소를 선택(인덱스 범위 지정)
print(sr.iloc[1 : 2])
print('\n')
print(sr.loc['생년월일' : '학생여부'])
````

- 시리즈 객체를 만드는 특수한 방법

````python
# <예제 1-3>
print(pd.Series())
print(pd.Series(5))
print(pd.Series(5, index=["a", "b", "c"]))
````

### 2-2 데이터프레임

데이터프레임은 2차원 배열이다. 열과 행으로 이루어져 있다.  
데이터프레임의 열은 각각 시리즈 객체이다. 시리즈를 열벡터(vector)라고 하면 데이터프레임은 여러개의 열벡터들이 같은 행 인덱스를 기준으로 줄지어 결합된 2차원 벡터 또는 행렬(matrix)이다.  
데이터프레임은 행 인덱스(row index)와 열이름(column name 또는 column label)으로 구분한다.  
데이터프레임의 열은 공통의 속성을 갖는 일련의 데이터를 나타내고, 행은 개별 관측대상에 대한 다양한 속성 데이터들의 모음인 레코드(record)가 된다.

- 데이터프레임 만들기

데이터프레임은 여러 개의 시리즈(열, column)을 모아 놓은 집합으로 이해하면 된다.  
데이터프레임을 만들 때는 판다스 DataFrame() 함수를 사용한다.  
`딕셔너리 -> 데이터프레임 변환: pandas.DataFrame( 딕셔너리 객체 )`

- 행 인덱스/열 이름 설정

2차원 배열에는 여러 개의 리스트(튜플)를 원소로 갖는 리스트(튜플), numpy 배열 등이 있다.  
2차원 배열을 DataFrame() 함수 인자로 전달하여 데이터프레임으로 변환할 때 행 인덱스와 열 이름 속성을 사용자가 직접 지정할 수도 있다.

````python
# <예제 1-4> 딕셔너리 -> 데이터프레임 변환
import pandas as pd

# 열이름을 key로 하고, 리스트를 value로 갖는 딕셔너리 정의(2차원 배열)
dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}

# 판다스 DataFrame() 함수로 딕셔너리를 데이터프레임으로 변환. 변수 df에 저장
df = pd.DataFrame(dict_data)

# df의 자료형 출력
print(type(df))

# 변수 df에 저장되어 있는 데이터프레임 객체를 출력
print(df)

# 주피터 환경에서 변수 출력
df
````

````python
# <예제 1-5> 행 인덱스/열 이름 설정
import pandas as pd

# 행 인덱스/열 이름 지정하여, 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']],
                  index=['준석', '예은'],
                  columns=['나이', '성별', '학교'])

# 데이터프레임 출력하기
print(df)

# 행 인덱스 확인하기
print(df.index)

# 열 이름 확인하기
print(df.columns)

# 행 인덱스 열 이름 변경하기
df.index=['학생1', '학생2']
df.columns=['연령', '남녀', '소속']

print(df)
print('\n')
print(df.index)
print('\n')
print(df.columns)
````

````python
# <예제 1-6> 행 인덱스/열 이름 변경
import pandas as pd

# 행 인덱스/ 열 이름 지정하여 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']],
                  index=['준석', '예은'],
                  columns=['나이', '성별', '학교'])

# 행 인덱스, 열 이름 확인하기
print(df)
print('\n')

# 열 이름 중, '나이'를 '연령'으로, '성별'을 '남녀'로, '학교'를 '소속'으로 바꾸기
df = df.rename(columns={'나이':'연령', '성별':'남녀', '학교':'소속'})

# df 출력(변경 후)
print(df)
print('\n')

# df의 행 인덱스 중에서, '준서'를 '학생1'로, '예은'을 '학생2'로 바꾸기
df = df.rename(index={'준석':'학생1', '예은':'학생2'})

# df 출력
print(df)
print('\n')
````

- 행/열 삭제

 데이터프레임의 행 또는 열을 삭제하는 명령으로 drop() 메소드가 있다. 행을 삭제할 때는 축(axis) 옵션으로 axis=0 또는 axis='index'를 입력하거나, 별도로 입력하지 않는다. 반면, 축  
옵션으로 axis=1 또는 axis='columns'를 입력하면 열을 삭제한다. 동시에 여러 개의 행 또는 열을 삭제하려면 리스트 형태로 함수 인자에 입력한다.  
한편, drop() 메소드는 기존 객체를 변경하지 않고 새로운 객체를 반환한다는 점에 유의한다. 따라서 원본 객체를 직접 변경하기 위해서는, 원본을 담고 있는 변수에 재할당하는 과정을 추가 한다.

````python
# <예제 1-7>행 삭제
import pandas as pd

# DataFrame() 함수로 데이터프레임 변환, 변수 df에 저장
exam_data = {'수학' : [90, 80, 70], '영어' : [98, 89, 95],
             '음악' : [85, 95, 100], '체육' : [100, 90, 90]}

df = pd.DataFrame(exam_data, index=['서준', '우현', '인아'])
print(df)
print('\n')

# 데이터프레임 df를 복제하여 변수 df2에 저장, df2의 1개 행(row) 삭제
df2 = df.copy()
df2 = df2.drop('우현')
print(df2)
print('\n')

# 데이터프레임 df를 복제하여 변수 df3에 저장, df3의 2개 행(row) 삭제
df3 = df.copy()
df3 = df3.drop(['서준', '인아'], axis=0)
print(df3) 
print('\n')

# 데이터프레임 df를 복제하여 변수 df4에 저장. df4의 2개 행(row) 삭제
df4 = df.copy()
df4 = df4.drop(['서준', '인아'], axis='index')
print(df4) 
print('\n')

# 데이터프레임 df를 복제하여 변수 df5에 저장. df5의 1개 행(row) 삭제
df5 = df.copy()
df5 = df5.drop(index=['우현'])
print(df5) 
print('\n')

````

````python
# <예제 1-8>열 삭제
import pandas as pd

# DataFrame() 함수로 데이터프레임 변환, 변수 df에 저장
exam_data = {'수학' : [90,80,70], '영어' : [98, 89, 95],
             '음악' : [85,95,100], '체육':[100,90,90]}

df = pd.DataFrame(exam_data, index=['서준', '우현', '인아'])
print(df)
print('\n')

# 데이터프레임 df를 복제하여 변수 df2에 저장. df2 1개 열(clumn) 삭제
df2 = df.copy()
df2 = df2.drop('수학', axis=1)
print(df2)
print('\n')

# 데이터프레임 df를 복제하여 변수 df3에 저장. df3 2개 열(clumn) 삭제
df3 = df.copy()
df3 = df3.drop(['영어', '음악'], axis=1)
print(df3)
print('\n')

# 데이터프레임 df를 복제하여 변수 df4에 저장. df4 2개 열(clumn) 삭제
df4 = df.copy()
df4 = df4.drop(['영어', '음악'], axis='columns')
print(df4)
print('\n')

# 데이터프레임 df를 복제하여 변수 df5에 저장. df5 1개 열(clumn) 삭제
df5 = df.copy()
df5 = df5.drop(columns=['수학'])
print(df5)
print('\n')
````


- 행 선택

데이터프레임의 행 데이터를 선택하기 위해서는 loc과 iloc 인덱서를 사용한다. 인덱스 이름을 기준으로 행을 선택할 때는 loc을 이용하고, 정수형 위치 인덱스를 사용할 때는 iloc을 이용한다.

````python
# <예제 1-9>행 선택
import pandas as pd

# DataFrame() 함수로 데이터프레임 변환, 변수 df에 저장
exam_data = {
    "수학": [90, 80, 70],
    "영어": [98, 89, 95],
    "음악": [85, 95, 100],
    "체육": [100, 90, 90],
}

df = pd.DataFrame(exam_data, index=["서준", "우현", "인아"])
print(df)
print("\n")

# 행 인덱스를 사용하여 행 1개 선택
label1 = df.loc['서준']     # loc 인덱스 활용
position1 = df.iloc[0]        # iloc 인덱스 활용
print(label1)
print('\n')
print(position1)
print('\n')
print(type(position1))
print('\n')

# 행 인덱스를 사용하여 2개 이상의 행 선택
label2 = df.loc[['서준', '우현']]     # loc 인덱스 활용
position2 = df.iloc[[0, 1]]        # iloc 인덱스 활용
print(label2)
print('\n')
print(type(label2))
print('\n')
print(position2)
print('\n')
print(type(position2))
print('\n')

# 행 인덱스의 범위를 지정하여 행 선택
label3 = df.loc['서준':'우현']     # loc 인덱스 활용
position3 = df.iloc[0:1]        # iloc 인덱스 활용
print(label3)
print('\n')
print(type(label3))
print('\n')
print(position3)
print('\n')
print(type(position3))
print('\n')

````


- 열 선택

데이터프레임에서 1개의 열을 선택하면 시리즈 객체가 반환된다. 데이터프레임의 열을 선택하여기 위해서는 대괄호([])안에 열 이름을 따옴표와 함께 입력하거나, 도트(.) 다음에 열 이름을 입력하는 두 가지 방식을 사용한다.

````python
# <예제 1-10>열 선택
import pandas as pd

# DataFrame() 함수로 데이터프레임 변환, 변수 df에 저장
exam_data = {'이름' : ["서준", "우현", "인아"],
             "수학": [90, 80, 70],
             "영어": [98, 89, 95],
             "음악": [85, 95, 100],
             "체육": [100, 90, 90],}

df = pd.DataFrame(exam_data)
print(df)
print("\n")
print(type(df))
print("\n")

# '수학' 점수 데이터만 선택, 변수 math1에 저장
math1 = df['수학']
print(math1)
print('\n')
print(type(math1))
print('\n')

# '영어' 점수 데이터만 선택, 변수 english에 저장
english = df["영어"]
print(english)
print('\n')
print(type(english))
print('\n')

# '음악', '체육' 점수 데이터를 선택, 변수 music_gym에 저장
music_gym = df[["음악", "체육"]]
print(music_gym)
print('\n')
print(type(music_gym))
print('\n')

# '수학' 점수 데이터만 선택, 변수 math2에 저장
math2 = df[["수학"]]
print(math2)
print('\n')
print(type(math2))
print('\n')

# 범위 슬라이싱의 고급 활용
print(df.iloc[::2])
print('\n')
print(df.iloc[:3:2])
print('\n')
print(df.iloc[::-1])
print('\n')
````


- 원소 선택

데이터프레임의 원소를 선택하기 위해 행 인덱스와 열 이름을 [행, 열] 형식의 2차원 좌표로 입력하여 원소 위치를 지정하는 방법이다. 원소가 위치하는 행과 열의 좌표를 입력하면 해당 위치의 원소가 반환된다.  
1개의 행과 2개 이상의 열을 선택하거나 반대로 2개 이상의 행과 1개의 열을 선택하는 경우 시리즈 객체가 반환된다. 2개 이상의 행과 2개 이상의 열을 선택하면 데이터프레임 객체를 반환한다.  


````python
# <예제 1-11>원소 선택
import pandas as pd

# DataFrame() 함수로 데이터프레임 변환, 변수 df에 저장
exam_data = {
    "이름": ["서준", "우현", "인아"],
    "수학": [90, 80, 70],
    "영어": [98, 89, 95],
    "음악": [85, 95, 100],
    "체육": [100, 90, 90],}
df = pd.DataFrame(exam_data)

# '이름' 열을 새로운 인덱스로 지정하고, df 객체에 변경 사항 반영
df = df.set_index('이름')
print(df)
print("\n")

# 데이터프레임 df의 특정 원소 1개 선택('서준'의 '음악' 점수)
a = df.loc['서준', '음악']
print(a)
print('\n')
b = df.iloc[0, 2]
print(b)
print('\n')

# 데이터프레임 df의 특정 원소 2개 이상 선택('서준'의 '음악', '체육' 점수)
c = df.loc['서준', ['음악','체육']]
print(c)
print('\n')
d = df.iloc[0, [2,3]]
print(d)
print('\n')
e = df.loc["서준", "음악" : "체육"]
print(e)
print('\n')
f = df.iloc[0, 2:]
print(f)
print('\n')

# df 2개 이상의 행과 열에 속하는 원소들 선택('서준', '우현'의 '음악', '체육' 점수)
g = df.loc[['서준', '우현'], ['음악','체육']]
print(g)
print('\n')
h = df.iloc[[0, 1], [2, 3]]
print(h)
print('\n')
i = df.loc['서준':'우현', '음악':'체육']
print(i)
print('\n')
j = df.iloc[0:2, 2:]
print(j)
print('\n')
````

벡터 자체가 배열이므로, 기존 행을 복사해서 새로운 행에 그대로 추가할 수도 있다.

`행 추가: DataFrame.loc['새로운 행 이름'] = 데이터 값 (또는 배열)`

- 행 인덱스 재배열

행 인덱스를 재배열하는 방법은 아래와 같다.  
`새로운 배열로 행 인덱스를 재지정: DataFrame 객체.reindex(새로운 인덱스 배열)`

````python
# <예제 1-17>새로운 배열로 인덱스를 재지정
import pandas as pd

# 딕셔너리를 데이터프레임으로 변환
dict_data = {
    "c0": [1,2,3],
    "c1": [4,5,6],
    "c2": [7,8,9],
    "c3": [10,11,12],
    "c4": [13,14,15]
}

# 인덱스를 [r0, r1, r2]로 지정
df = pd.DataFrame(
    dict_data,
    index=[
        "r0",
        "r1",
        "r2",
    ],
)
print(df)
print("\n")

# 인덱스를 [r0, r1, r2, r3, r4]로 재지정
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf = df.reindex(new_index)
print(ndf)
print("\n")

# reindex로 발생한 NaN 갑을 숫자 0으로 채우기
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf2 = df.reindex(new_index, fill_value=0)
print(ndf2)
print("\n")
````

```행 인덱스 기준 정렬: DataFrame 객체.sort_index()

행 인덱스를 기준으로 데이터프레임을 정렬하는 방법은 아래와 같다.  
`행 인덱스 기준 정렬: DataFrame 객체.sort_index()`

````python
# <예제 1-19>데이터프레임 정렬
import pandas as pd

# 딕셔너리를 데이터프레임으로 변환
dict_data = {
    "c0": [1, 2, 3],
    "c1": [4, 5, 6],
    "c2": [7, 8, 9],
    "c3": [10, 11, 12],
    "c4": [13, 14, 15],
}

# 인덱스를 [r0, r2, r1]로 지정
df = pd.DataFrame(
    dict_data,
    index=[
        "r0",
        "r2",
        "r1",
    ],
)
print(df)
print("\n")

# 내림차순으로 행 인덱스 정렬
ndf = df.sort_index(ascending=False)
print(ndf)
print("\n")

# 오름차순으로 행 인덱스 정렬
ndf2 = df.sort_index(ascending=True)
print(ndf2)
print("\n")
````

- 특정 열의 데이터 값을 기준으로 데이터프레임 정렬하기

특정 열의 데이터 값을 기준으로 데이터프레임을 정렬하는 방법은 아래와 같다.

````python
# <예제 1-20>열 기준 정렬
import pandas as pd

# 딕셔너리를 데이터프레임으로 변환
dict_data = {
    "c0": [1, 2, 3],
    "c1": [4, 5, 6],
    "c2": [7, 8, 9],
    "c3": [10, 10, 11],
    "c4": [15, 14, 14],
}

# 인덱스를 [r0, r2, r1]로 지정
df = pd.DataFrame(
    dict_data,
    index=[
        "r0",
        "r2",
        "r1",
    ],
)
print(df)
print("\n")

# c1 열을 기준으로 내림차순 정렬
ndf = df.sort_values(by='c1', ascending=False)
print(ndf)
print("\n")

# c2 열을 기준으로 오름차순 정렬
ndf2 = df.sort_values(by='c1', ascending=True)
print(ndf2)
print("\n")

# c3 열을 기준으로 내림차순 정렬, c4 열을 기준으로 오름차순 정렬
ndf3 = df.sort_values(by=['c3', 'c4'], ascending=[False, True])
print(ndf3)
print("\n")
````

- 시리즈 vs 시리즈

시리즈와 시리즈 사이의 사칙연산을 처리하는 방법은 아래와 같다.  
`시리즈와 시리즈 연산: Series1 + 연산자(+,-,*,/) + Series2`

````python
# <예제 1-22> 시리즈 사칙연산
# 라이브러리 불러오기
import pandas as pd

# 딕셔너리 데이터로 판다스 시리즈 만들기
student1 = pd.Series({"국어": 100, "영어": 80, "수학": 90})
student2 = pd.Series({"수학": 80, "국어": 100, "영어": 80})
print(student1)
print("\n")
print(student2)

# 두 학생의 과목별 정수로 사칙연산 수행
addition = student1 + student2      # 곱셈
print(addition)
print(type(addition))
print("\n")

subtraction = student1 - student2      # 뺄셈
print(subtraction)
print(type(subtraction))
print("\n")

multiplication = student1 * student2      # 곱셈
print(multiplication)
print(type(multiplication))
print("\n")

division = student1 / student2      # 나눗셈
print(division)
print(type(division))
print("\n")

# 사칙연산 결과를 데이터프레임으로 합치기(시리즈 -> 데이터프레임)
result = pd.DataFrame([addition, subtraction, multiplication, division],
                      index=['덧셈', '뺄셈', '곱셈', '나눗셈'])

print(result)
````

````python
# <예제 1-23>NaN 값이 있는 시리즈 연산
# 라이브러리 불러오기
import pandas as pd
import numpy as np

# 딕셔너리 데이터로 판다스 시리즈 만들기
student1 = pd.Series({"국어": np.nan, "영어": 80, "수학": 90})
student2 = pd.Series({"수학": 80, "국어": 100})
print(student1)
print("\n")
print(student2)

# 두 학생의 과목별 점수로 사칙연산 수행(시리즈 vs 시리즈)
addition = student1 + student2          # 덧셈
subtraction = student1 - student2          # 뺄셈
multiplication = student1 * student2          # 곱셈
division = student1 / student2          # 나눗셈
print(type(division))
print("\n")

# 사칙연산 결과를 데이터프레임으로 합치기(시리즈 -> 데이터프레임)
result = pd.DataFrame(
    [addition, subtraction, multiplication, division],
    index=[
        "덧셈",
        "뺄셈",
        "곱셈",
        "나눗셈",
    ],
)
print(result)
````

- 연산 메소드

객체 사이에 공통 인덱스가 없어서 NaN값이 나오는데 이름 0으로 나오게 하는 방법  
`연산 메소드 사용(시리즈와 시리즈 덧셈): Series.add(Series2, fill_value=0)`

````python
# <예제 1-24>연산 메소드 사용 - 시리즈 연산
# 라이브러리 불러오기
import pandas as pd
import numpy as np

# 딕셔너리 데이터로 판다스 시리즈 만들기
student1 = pd.Series({"국어": np.nan, "영어": 80, "수학": 90})
student2 = pd.Series({"수학": 80, "국어": 90})
print(student1)
print("\n")
print(student2)
print("\n")

# 두 학생의 과목별 점수로 사칙연산 수행(시리즈 vs 시리즈)
sr_add = student1.add(student2, fill_value=0)  # 덧셈
sr_sub = student1.sub(student2, fill_value=0)  # 뺄셈
sr_mul = student1.mul(student2, fill_value=0)  # 곱셈
sr_div = student1.div(student2, fill_value=0)  # 나눗셈
print(type(division))
print("\n")

# 사칙연산 결과를 데이터프레임으로 합치기(시리브 -> 데이터프레임)
result = pd.DataFrame(
    [sr_add, sr_sub, sr_mul, sr_div],
    index=[
        "덧셈",
        "뺄셈",
        "곱셈",
        "나눗셈",
    ],
)
print(result)
````

### 4-2 데이터프레임 연산

행/열 인덱스를 기준으로 일대일 대응되는 원소끼리 연산을 처리한다.

- 데이터프레임 vs 숫자

데이터프레임에 숫자을 연산하는 방법은 다음과 같다.  
`데이터프레임과 숫자 연산: DataFrame 객체 + 연산자(+,-,*,/) + 숫자`

````python
# <예제 1-25> 데이터프레임에 숫자 더하기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'fare']]
print(df.head())
print('\n')
print(type(df))

# 데이터프레임에 숫자 10 더하기
addition = df + 10
print(addition.head())
print('\n')
print(type(addition))
````

- 데이터프레임 vs 데이터프레임

데이터프레임간에 연산하는 방법은 아래와 같다.  
`데이터프레임의 연산자 활용: DataFrame1 + 연산자(+,-,*,/) + DataFrame2`

````python
# <예제 1-26> 데이터프레임끼리 더하기
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "fare"]]
print(df.tail())
print("\n")
print(type(df))
print("\n")

# 데이터프레임에 숫자 10 더하기
addition = df + 10
print(addition.tail())
print("\n")
print(type(addition))
print("\n")

# 데이터프레임끼리 연산하기(addition -df)
subtraction = addition - df
print(subtraction.tail())
print("\n")
print(type(subtraction))
print("\n")

# 사칙연산 수행(연산 메소드 사용)
sample1 = addition.tail()
sample2 = subtraction.tail().fillna(0.0)
print(sample1)
print("\n")
print(sample2)
print("\n")

df_add = sample1.add(sample2, fill_value=0)     #덧셈
df_sub = sample1.sub(sample2, fill_value=0)     #뺄셈
df_mul = sample1.mul(sample2, fill_value=0)     #곱셈
df_div = sample1.div(sample2, fill_value=0)     #나눗셈

print(df_add)
print('\n')
print(df_sub)
print('\n')
print(df_mul)
print('\n')
print(df_div)
print('\n')
````

## 필터링

시리즈 또는 데이터프레임의 데이터 중에서 특정 조건식을 만족하는 원소만 따로 추출하는 개념이다. 가장 대표적인 방법인 불린 인덱싱(boolean indexing)을 알아보자.


### 5-1 불린 인덱싱

불린 인덱싱을 하는 방법은 아래와 같다.  
`데이터프레임의 불린 인덱싱: DataFrame객체[불린 시리즈]`

````python
# <예제 1-27> 불린 인덱싱
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋로딩
titanic = sns.load_dataset("titanic")

# 행 인덱스 0-9 범위에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
df = titanic.loc[0:9, ["age", "fare"]]
print(df)
print("\n")

# 조건식 적용(20세 미만: age < 20)
print(df['age'] < 20)

# 불린 인덱싱(20세 미만의 조건식을 충족하는 행을 필터링)
print(df[df['age']<20])

# 불린 인덱싱(loc 인덱서 활용)
print(df.loc[df['age']<20])

# 불린 인덱싱(NOT 논리연산)
print(df.loc[~(df['age']<20)])

# 나이가 10대(10~19세)인 승객(AND 논리연산)
mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic[mask1]
print(df_teenage.head())

# 나이가 10세 미만(0~9세)이고 여성인 승객(AND 논리연산, loc 인덱서)
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic[mask2]
print(df_female_under10.head())

# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객(OR 논리연산)
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic[mask3]
print(df_under10_morethan60.head())

# 불린 시리즈를 새로운 열로 추가
df['age_under_20'] = df['age'] < 20
print(df.head())
````

***판다스에서 불린 자료형의 활용***  
판다스에서 불린 자료형은 데이터 처리와 분석에서 다양하게 활용된다.  
- 조건 필터링: 불린 인덱싱을 사용하여 특정 조건을 만족하는 행을 필터링할 수 있다.  
- 새로운 열 생성: 조건에 따라 새로운 불린 열을 생성하여 데이터프레임에 추가할 수 있다. 

### 5-2 query() 메소드 활용

query()메소드 사용방법은 아래와 같다.  
`query 메소드를 활용한 필터링: DataFrame객체.query(조건 문자열)`

````python
# <예제 1-28> query 필터링
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋로딩
titanic = sns.load_dataset("titanic")

# 행 인덱스 0-9 범위에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
df = titanic.loc[0:9, ["age", "fare"]]
print(df)
print("\n")

# 20세 미만의 조건을 충족하는 행을 필터링
print(df.query('age < 20'))
print("\n")

# 나이가 20세 이상이고 요금이 15 미만인 조건을 모두 충족하는 행을 필터링
print(df.query('age >= 20 and fare < 15'))
print("\n")

# 나이가 20세 이상 또는 요금이 15 미만인 조건을 하나라도 충족하는 행을 필터링
print(df.query('age >= 20 or fare < 15'))
print("\n")
````

### 5-3 isin()메소드 활용

isin()메소드를 활용하는 방법은 아래와 같다.  
`isin()메소드를 활용한 필터링: DataFrame의 열 객체.isin(추출 값의 리스트)`

````python
# <예제 1-29> isin() 필터링
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋로딩
titanic = sns.load_dataset("titanic")

# 디스플레이 설정 변경 - 출력할 최대 열의 개수
pd.set_option('display.max_columns', 10)

# 탑승한 도시가 "Southampton", "Queenstown"인 승객만 따로 추출 - 불린 인덱싱
mask1 = titanic["embark_town"] == "Southampton"
mask2 = titanic["embark_town"] == "Queenstown"
df_boolean = titanic[mask1 | mask2]
print(df_boolean.head())

# isin() 메소드 활용하여 동일한 조건으로 추출
isin_filter = titanic["embark_town"].isin(["Southampton", "Queenstown"])
df_isin = titanic[isin_filter]
print(df_isin.head())
````

`StringDtype(string 자료형)`

````python
# <예제 1-30> 텍스트 자료형
# 라이브러리 불러오기
import pandas as pd

# 텍스트로 이루어진 시리즈 배열 만들기(자료형 미지정) -> 아무래도 업데이트 되면서 시리즈 객체안의 원소가 모두 문자열이면 str타입으로 나오는 듯 함....
fruit_name = pd.Series(['Apple', 'Banana', 1])

# 시리즈 출력
print(fruit_name)
print('\n')

# 자료형 저장: 'string' 사용
print(pd.Series(["Apple", "Banana", 1], dtype="string"))
print('\n')

# 자료형 지정: pd.StringDtype() 사용
print(pd.Series(["Apple", "Banana", 1], dtype=pd.StringDtype))
print('\n')

# 자료형 변환
print(fruit_name.astype('string'))
 
````

### 6-2 문자열 메소드(string methods)

판다스의 1차원 배열 객체인 시리즈와 인덱스는 각 요소에 대해 작업을 쉽게 수행할 수 있는 다양한 문자열 처리 메소드를 제공한다.  
`문자열 메소드 적용(lower): Series 객체.str.lower() 또는 Index 객체.str.lower()`

````python
# <예제 1-31> 문자열 메소드
# 라이브러리 불러오기
import pandas as pd
import numpy as np

# 텍스트로 이루어진 시리즈 배열 만들기(자료형 미지정) -> 아무래도 업데이트 되면서 시리즈 객체안의 원소가 모두 문자열이면 str타입으로 나오는 듯 함....
ser = pd.Series(["Apple_사과", "Banana_바나나", "cherry_체리", np.nan],
                       index=["First", "Second", "Third", "Fourth"])

print(ser)
print("\n")

# 텍스트로 이루어진 시리즈 배열 만들기(string 자료형 지정)
ser2 = pd.Series(["Apple_사과", "Banana_바나나", "cherry_체리", np.nan],
                       index=["First", "Second", "Third", "Fourth"], dtype="string")
print(ser2)
print("\n")

# 소문자로 변환
print(ser.str.lower())
print("\n")

# 대문자로 변환
print(ser.str.upper())
print("\n")

# 문자열 길이 계산
print(ser.str.len())
print("\n")

# 문자열 분할
print(ser.str.split("_"))
print("\n")

# 문자열을 분할하여 DataFrame으로 확장
print(ser.str.split("_", expand=True))
print("\n")

# 분할된 문자열 중 특정 요소 선택
print(ser.str.split("_").str.get(1))
print("\n")

# 행 인덱스를 추출
idx = ser.index
print(idx)
print("\n")

# 양쪽 공백 제거
print(idx.str.strip())
print("\n")

# 왼쪽 공백 제거
print(idx.str.lstrip())
print("\n")

# 오른쪽 공백 제거
print(idx.str.rstrip())
print("\n")

# 문자열 교체("_"을 찾아서 ":"으로 변경)
print(ser.str.replace("_", ":", regex=False))
print("\n")

# 정규 표현식을 사용한 문자열 교체(영문 알파벳이 아닌 문자 제거)
print(ser.str.replace("[^a-zA-Z\s]", "", regex=True))
print("\n")

# 인덱싱 - 각 문자열의 첫 번째 문자 선택
print(ser.str[0])
print("\n")

# 슬라이싱 - 각 문자열의 첫 번째부터 네 번째 문자 선택
print(ser.str[0:4])
print("\n")

# 특정 문자 포함 여부 확인, NA 같은 값을 False로 처리
contains_A = ser.str.contains("A", na=False)
print(contains_A)
print("\n")

# 패턴 포함 여부 확인
contains_pattern = ser.str.contains(r"[A|B][a-z]+")
print(contains_pattern)
print("\n")

# 전체 문자열의 패턴 일치 여부 확인
fullmatch_pattern = ser.str.fullmatch(r"[A|B][a-z]+")
print(fullmatch_pattern)
print("\n")

# 새로운 시리즈 생성
ser3 = pd.Series(["a1", "b2", "c3", "d4a5", "e6e7", "f8"], dtype="string")

# 숫자만 추출
digits = ser3.str.extractall(r"(\d)").groupby(level=0).agg(''.join)
print(digits)
print("\n")

# 알파벳만 추출
letters = ser3.str.extractall(r"([a-z])").groupby(level=0).agg(''.join)
print(letters)
print("\n")

# 숫자를 별도의 컬럼으로 분리
split_columns = ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Digit>\d+)")
print(split_columns)
````
